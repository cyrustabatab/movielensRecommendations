```python
import pandas as pd
```

##### In the users.dat file, there are four columns: user id, gender, age, occupation, and zip code. For age, the numbers are chosen from (1,18,25,35,45,50,56) with each of those numbers mapping to a range. Similarly, there are 21 occupations with each number from 0-20 corresponding to an occupation. For easy conversion, the following cell provides a mapping between the numbers and a human readable string for the age and occupation columns. The descriptions and ranges are taken straight from the movielens website(http://files.grouplens.org/datasets/movielens/ml-1m-README.txt)

```python
#Create mapping between number in age and occupation column to human readable string

ages = {1: "<18", 18: "18-24", 25: "25-34", 35: "35-44", 45: "45-49", 50: "50-55"
       ,56: "56+"}


occupations = {0: "other/not specified", 1: "academic/educator", 2: 'artist'
              ,3: "clerical/admin", 4: "college/grad student", 5: "customer service"
              ,6: "doctor/health care", 7: "executive/managerial", 8: "farmer", 9: "homemaker"
              ,10: "K-12 student", 11: "lawyer", 12: "programmer", 13: "retired", 14: "sales/marketing"
              ,15: "scientist", 16: "self-employed", 17: "technician/engineer", 18: "tradesman/craftsman"
              ,19: "unemployed", 20: "writer"}
```

```python
#Get the ratings file and save to ratings.csv

ratings = pd.read_csv('data/ratings.dat',sep="::",engine='python',encoding='latin-1',names=['user_id','movie_id','rating','timestamp'])

#ratings['user_embedded_id'] = ratings['user_id'] - 1 #Needed for Keras
#ratings['movie_embedded_id'] = ratings['movie_id'] - 1 #Needed for Keras

#Save to CSV file
ratings.to_csv("ratings.csv", sep='\t',header=True,columns=['user_id', 'movie_id', 'rating','timestamp'])
```

```python
#Get the users file and save to users.csv

users = pd.read_csv('data/users.dat',sep="::",engine='python',encoding='latin-1',names=['user_id','gender','age','occupation','zipcode'])

users['age_description'] = users['age'].apply(lambda x: ages[x])
users['occupation_description'] = users['occupation'].apply(lambda x: occupations[x])

users.to_csv("users.csv",sep='\t',header=True,columns=['user_id','gender','age','occupation','zipcode', 'age_description', 'occupation_description'])

print(len(users))
```

```python
movies = pd.read_csv("data/movies.dat",sep="::",engine='python',encoding='latin-1',names=['movie_id', 'title', 'genres'])

movies.to_csv("movies.csv", sep='\t',header=True,columns=['movie_id','title','genres'])

print(len(movies))
```
